```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('./ZestyAI-combined_structure_status.csv')
```

```python
d.head()
```

```python
d['DAMAGE'].value_counts().plot(kind='bar')
```

```python
d.loc[d['DAMAGE']=='Destroyed (>50%)','DAMAGE'] = 'Destroyed'
d.loc[d['DAMAGE']=='Affected (1-9%)','DAMAGE'] = 'Affected'
d.loc[d['DAMAGE']=='1-9%','DAMAGE'] = 'Affected'
d.loc[d['DAMAGE']=='Minor (10-25%)','DAMAGE'] = 'Minor'
d.loc[d['DAMAGE']=='10-25%','DAMAGE'] = 'Minor'
d.loc[d['DAMAGE']=='Major (26-50%)','DAMAGE'] = 'Major'
d.loc[d['DAMAGE']=='26-50%','DAMAGE'] = 'Major'
d.loc[d['DAMAGE']=='51-75%','DAMAGE'] = 'Destroyed'
```

```python
d['DAMAGE'].value_counts().plot(kind='bar')
```

```python
plt.scatter(d.DAMAGE, d.CLAIM)
```

```python
d['VEGCLEARAN'].value_counts().plot(kind='bar')
```

```python
d.loc[d['VEGCLEARAN']=="0-30'",'VEGCLEARAN'] = '0-30'
d.loc[d['VEGCLEARAN']=="30-100'",'VEGCLEARAN'] = '30-100'
d.loc[d['VEGCLEARAN']==">100'",'VEGCLEARAN'] = '>100'
d.loc[d['VEGCLEARAN']=="Unknown",'VEGCLEARAN'] = 'unknown'
d.loc[d['VEGCLEARAN']=="-",'VEGCLEARAN'] = 'unknown'
d['VEGCLEARAN'].value_counts().plot(kind='bar')
```
